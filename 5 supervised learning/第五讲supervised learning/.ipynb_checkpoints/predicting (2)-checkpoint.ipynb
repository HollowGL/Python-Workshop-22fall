```python
import pandas as pd
import jieba.posseg as pseg
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import NMF
import requests
from sklearn.cross_validation import train_test_split, cross_val_score
import numpy as np
import math
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.multiclass import OneVsRestClassifier
import matplotlib.pyplot as plt
from sklearn.naive_bayes import BernoulliNB
from sklearn.svm import SVC
from sklearn.metrics import roc_curve

import os

%matplotlib inline
```

```python
#载入定义文本清洗函数

stopwords_url = 'https://raw.githubusercontent.com/Isac-Cooper/chinese-stopwords/master/stopwords'
stopwords = requests.get(stopwords_url).text.split()
cixing = ["x","zg","uj","ul","e","d","uz","y","m","eng",'ns','nr']
def cut(txt):
    result = ''
    try:
        for w in pseg.cut(txt):
            if not str(w.flag) in cixing:
                seg=w.word
                if seg not in stopwords:
                    result += seg + ' '
    except:
        pass
    return result
```

```python
#读取训练集并进行清洗
codegroupA = pd.read_excel('trainingset.xlsx')
try:
    codegroupA['parsed']
except:
    codegroupA['parsed'] = codegroupA.content.apply(cut)
```

```python
#学习话题模型 with multilabelbinarizer
topics = []
for index, row in codegroupA.iterrows():
    topic = []
    for field in ['topic' + str(a) for a in range(1,3)]:
        if not pd.isnull(row[field]):
            topic.append(row[field])
    if topic == []:
        topic = [0.0]
    topics.append(topic)
    
mlb = MultiLabelBinarizer()
y = mlb.fit_transform(topics)
y.shape

model = OneVsRestClassifier(LogisticRegression())
model.fit(X,y)
```

```python
for model_ in [RandomForestClassifier(), LogisticRegression(), SVC(), BernoulliNB()]:
    print(cross_val_score(OneVsRestClassifier(model_), X, y))
```

```python
#预测话题模型
predict = model.predict(dtm_full)
np.save('topic',model.predict(dtm_full))
full['predict_topic'] = mlb.inverse_transform(predict)
```

```python
#学习身份模型
codegroupA['status1'] = codegroupA['status1'].replace('',np.nan).replace(' ',np.nan).astype('float')
codegroupA['status2'] = codegroupA['status2'].replace('',np.nan).replace(' ',np.nan).astype('float')
codegroupA['status3'] = codegroupA['status3'].replace('',np.nan).replace(' ',np.nan).astype('float')
statuses = []
for index, row in codegroupA.iterrows():
    status = []
    for field in ['status' + str(a) for a in range(1,4)]:
        if not pd.isnull(row[field]):
            status.append(row[field])
    if status == []:
        status = [0.0]
    statuses.append(status)
    
mlb = MultiLabelBinarizer()
y = mlb.fit_transform(statuses)
y.shape

model = OneVsRestClassifier(LogisticRegression())
model.fit(X,y)
```
