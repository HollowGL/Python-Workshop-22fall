```python
#import os
#import re
import pandas as pd
import numpy as np
import jieba.posseg as pseg
from sklearn.decomposition import PCA
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import jieba
import matplotlib.pyplot as plt
%matplotlib inline
```

## PCA

```python
from sklearn.datasets import load_breast_cancer
```

```python
breast = load_breast_cancer()
breast_data = breast.data
breast_data.shape
```

```python
breast_dataset = pd.DataFrame(breast_data)
features = breast.feature_names
features
```

```python
breast_dataset.columns = features
breast_dataset.head()
```

```python
from sklearn.preprocessing import StandardScaler #PCA 要求标准化
x = breast_dataset.loc[:, features].values
x = StandardScaler().fit_transform(x) # normalizing the features
np.mean(x),np.std(x)
```

```python
normalised_breast = pd.DataFrame(x,columns=features)
normalised_breast.head()
```

```python
pca_breast = PCA(n_components=2)
principalComponents_breast = pca_breast.fit_transform(x)
principal_breast_Df = pd.DataFrame(data = principalComponents_breast
             , columns = ['principal component 1', 'principal component 2'])
```

```python
principal_breast_Df
```

```python
print('Explained variation per principal component: {}'.format(pca_breast.explained_variance_ratio_))
```

```python
loadings = pca_breast.components_
num_pc = pca_breast.n_features_
pc_list = ["PC"+str(i) for i in list(range(1, num_pc+1))]
loadings_df = pd.DataFrame.from_dict(dict(zip(pc_list, loadings)))
loadings_df['variable'] = features
loadings_df = loadings_df.set_index('variable')
loadings_df
```

```python
plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principal Component - 1',fontsize=20)
plt.ylabel('Principal Component - 2',fontsize=20)
plt.title("Principal Component Analysis of Breast Cancer Dataset",fontsize=20)
targets = [0, 1]
colors = ['r', 'g']
for target, color in zip(targets,colors):
    indicesToKeep = breast.target == target
    plt.scatter(principal_breast_Df.loc[indicesToKeep, 'principal component 1']
               , principal_breast_Df.loc[indicesToKeep, 'principal component 2'], c = color, s = 50)

plt.legend(targets,prop={'size': 15})
```

## LDA topic modeling

```python
stopwords = [k.strip() for k in open('cn_stopwords.txt', encoding='utf8').readlines() if k.strip() != '']
```

```python
report= pd.read_csv('samgov.csv')
report.head()
```

```python
txt_list = []
cixing=["x","zg","uj","ul","e","d","uz","y","eng","m"]
for i, txt in enumerate(report['content']):
    result = ''
    try:
        for w in pseg.cut(txt):
                if not str(w.flag)  in cixing:
                    seg= w.word
                    if seg not in stopwords:
                            result+= seg +" "
    finally:
        pass
    txt_list.append(result)
```

```python
print(txt_list[2])
```

```python
cv = CountVectorizer(min_df=1)
dtm = cv.fit_transform(txt_list)
vocab = cv.get_feature_names_out()
```

```python
lda = LatentDirichletAllocation(n_components=15, max_iter=200,)
l = lda.fit_transform(dtm)
```

```python
print(l.shape)  # (no_of_doc,no_of_topics)
print(l[0])
```

```python
l.sum(axis = 1)
```

```python
# composition of doc 0 for eg
print("Document 0: ")
for i,topic in enumerate(l[0]):
  print("Topic ",i,": ",topic*100,"%")
```

```python
# most important words for each topic
def print_top_words(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print("Topic #%d:" % topic_idx)
        print(" ".join([feature_names[i]
                        for i in topic.argsort()[:-n_top_words - 1:-1]]))
    print()
print_top_words(lda, vocab,50)
```

```python
import pyLDAvis 
import pyLDAvis.sklearn
pyLDAvis.enable_notebook ()
pic = pyLDAvis.sklearn.prepare(lda,dtm,cv )
pyLDAvis.show(pic, local=False)
```

```python
plexs = []
scores = []
n_max_topics =16 #random set

for i in range(1,n_max_topics):
    lda = LatentDirichletAllocation(n_components=i, max_iter=50,learning_method='batch',
                                    learning_offset=50, random_state=666)

    lda.fit(dtm)
    plexs.append(lda.perplexity(dtm))
    scores.append(lda.score(dtm))

    
```

```python
n_t=15 # 区问最右侧的值。注意：不能大于n_max_topics
x=list(range (1,n_t))
plt.plot(x,plexs [1:n_t])
plt.xlabel("number of topics")
plt.ylabel("perplexity")
plt.show()
```

```python
n_t=15 # 区问最右侧的值。注意：不能大于n_max_topics
x=list(range (1,n_t))
plt.plot(x,scores [1:n_t])
plt.xlabel("number of topics")
plt.ylabel("score")
plt.show()
```
