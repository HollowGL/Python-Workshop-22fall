```python
# 如果想设计更加丰富多彩的图：

# 简单的绘图
# x = np.linspace(-5, 10, 50) #调用np.linspace是创建一个 numpy 数组，并记作 x。x 包含了从 -5 到 10 之间等间隔的 50 个值。
# print(x)

# y1 = x + 3      # 曲线 y1
y2 = 3 - x      # 曲线 y2


######## 开始画图 ########

plt.figure()    # 建立一个画布
plt.plot(x, y1, linestyle = '-.', linewidth = 3.0, color = '#660874') # 绘制曲线 y1 # linestyle 标注了数据，linewidth 宽度
plt.plot(x, y2, marker = '*', markersize = 8.0,  color = '#9999ff') # 绘制曲线 y2 # 用 marker 标注的数据点

# 设置横轴的上下限
plt.xlim(-6, 12)
# 设置纵轴的上下限
plt.ylim(-20, 60)

# 设置横轴标签
plt.xlabel("X's label")
# 设置纵轴标签
plt.ylabel("Y's label")

plt.show() # 呈现图片


### 选取配色代码的网站
###  https://www.cnblogs.com/kuoAT/p/7198794.html

# 清华紫的RGB是 RGB 102 8 116, HEX代码 是 #660874

```

```python
# 在一页上放两个图

fig, axes = plt.subplots(1, 2) # 建立一个1*2（一行，两列）的图, axes 可以随便改
y2 = 3 - x      # 画了一条曲线 y2


axes[0].plot(x, y1, color = 'red') #linewidth = 8.0
axes[0].set_title('y1 = x + 3') # 设置标题
axes[0].set_ylabel('common ylabel')
axes[1].plot(x, y2, color = 'blue')
axes[1].set_title('y2 = 3 - x')
axes[1].set_ylabel('common ylabel') # 还记得上面的设置横纵轴标签的问题吗？
fig.tight_layout()
plt.show() # 呈现图片



```

```python
# 如何绘制更多的图？

fig, axs = plt.subplots(2, 2, 
                        sharex = True) # try removing sharex = True


# 现在设置4条不同的函数曲线
y1 = x + 3      
y2 = 3 - x    
y3 = 2 * x
y4 = x ** 2

plt.figure(figsize = (20, 12))    # 建立一个画布
axs[0,0].plot(x, y1, color = 'green')
axs[0,0].set_title('y1 = x + 3', fontsize = 20) # 改变字体大小
axs[0,1].plot(x, y2)
axs[0,1].set_title('y2 = 3 - x', color = 'b') # 改变颜色
axs[1,0].plot(x, y3)
axs[1,0].set_title('y3 = 2 * x', weight = 'bold') # 改变字体粗细
axs[1,1].plot(x, y4)
axs[1,1].set_title('y4 = x ^ 2') # default setting
fig.tight_layout() 

plt.show() # 呈现图片









```

```python
# how to use different scale for the same axis
plt.figure(figsize = (8,8)) 
small = 12 # 设置一个字号值
medium = 15 # 设置一个字号值
plt.rc('axes', titlesize=small, labelsize=medium) # 设置轴
plt.rc('xtick', labelsize=small) # 设置x的scale   
plt.rc('ytick', labelsize=medium) # 设置y的scale


fig, ax1 = plt.subplots()
color = 'tab:blue' # Tableau Colors
ax1.set_xlabel('x')
ax1.set_ylabel('y = x + 3', color = color)
ax1.plot(x, y1, color=color)
ax1.tick_params(axis='y', 
                labelcolor=color)



ax2 = ax1.twinx()  # a very important step: second axes that shares the same x-axis

color = 'tab:pink'
ax2.set_ylabel('y = x ** 2', color=color)  # we already handled the x-label with ax1
ax2.plot(x, y4, color=color)
ax2.tick_params(axis='y', 
                labelcolor=color)
fig.suptitle('2*2 plots') #给图加标题 centered title

# fig.tight_layout()
plt.show()




```

```python
# 用matplotlib绘图时的中文问题
# 通常可以使用配置字体库的方式，见 https://blog.csdn.net/qq_17753903/article/details/86260276
# 但这个比较复杂，初学者可以使用以下方法进行设置



# 设置字体路径的方式 - 方便起见，可以直接放在存储jupyter notebook的文件夹中。
zhfont = matplotlib.font_manager.FontProperties(fname = r'SourceHanSerifCN-Medium.ttf')



# 前面使用过的数据
# x = np.linspace(-5, 10, 50) #调用np.linspace是创建一个 numpy 数组，并记作 x。x 包含了从 -5 到 10 之间等间隔的 50 个值。
# print(x)
# y1 = x + 3      # 曲线 y1
# y2 = 3 - x      # 曲线 y2

plt.figure()    # 建立一个画布
plt.plot(x, y1, linestyle = '--', color = '#660874', linewidth = 3.0) # 绘制曲线 y1 # linestyle 标注了数据，linewidth 宽度
plt.plot(x, y2, marker = 'D', color = '#9999ff', markersize = 4.0) # 绘制曲线 y2 # 用 marker 标注的数据点


# 设置横轴的上下限，刻度
plt.xlim(-4, 8)
# 设置纵轴的上下限，刻度
plt.ylim(-10, 20)

# 设置横轴标签
plt.xlabel("横轴中文标签", fontproperties=zhfont, size = 15) # 通过 font properties 对字体的属性进行设置
# 设置纵轴标签
plt.ylabel("纵轴中文标签", fontproperties=zhfont, size = 15, rotation = 75) # 默认 rotation = 90

# 设置图例
plt.legend(labels = ['X', 'Y'], loc = 'upper right', fontsize = 10)

# 绘制散点(3, 6)
plt.scatter([3], [6], s=30, color="blue")      # s 为点的 size
# 对(3, 6)做标注
plt.annotate("(3, 6)",
             xy=(3, 9),       # 在(3.3, 5.5)上做标注
             fontsize=16,         # 设置字体大小为 16
             xycoords='data')  # xycoords='data' 是说基于数据的值来选位置


# 对(3, 0)做标注
plt.text(2.3, 2.6, "this point very important", # 2.3, 2.6 是相对位置
         fontdict={'size': 12, 'color': 'blue'})  # fontdict设置文本字体


plt.show() # 呈现图片


```

```python
# seaborn use see https://seaborn.pydata.org/tutorial/regression.html
tips = sns.load_dataset("tips")
tips[:10]

# continuous and categorical variable
# total_bill 总账单
# tip 消费
# sex 性别
# smoker 是否吸烟
# day 日期（星期x）
# time （午饭/晚饭）
# size 人数
```

```python
# 如果报错：AttributeError: module 'seaborn' has no attribute 'histplot'，
# 在这里加上：pip install -U seaborn
# 然后重启 notebook


plt.figure(figsize = (20, 12)) 
fig, axes = plt.subplots(1, 2) # 1*2的格子

sns.histplot(tips, # dataset
            x = "size",
            hue ="time", # 颜色
            multiple ="dodge", # 尝试一下 multiple ="stack"
            discrete = True, # 尝试一下改成 False
            palette="rocket", # seaborn 自己的 palette # see https://seaborn.pydata.org/tutorial/color_palettes.html?highlight=palette
            bins=[1, 2, 3, 4, 5, 6],
            stat='count', #stat = "density" 
            ax=axes[0]) # 看一下，这里的

# 我们看到的是在某一餐，每桌有几个人的分布

sns.histplot(tips, 
            x = "size",
            hue ="time", 
            multiple ="stack", # 尝试一下 multiple ="stack"
            discrete = True, # are centered on their corresponding data points
            palette="rocket", # seaborn 自己的 palette
            bins=[1, 2, 3, 4, 5, 6],
            stat='count',
            ax=axes[1])


plt.text(4, 50, "hi this is annotation", #hi \nthis \nis \nannotation
         horizontalalignment='left', 
         fontsize = 10, 
         color='blue', 
         weight='bold') #light

plt.show()
```

```python
sns.lmplot(x = "size", # variable name：size of diners
           y = "tip", # variable name
           hue = "smoker", # condition on other variables
           data = tips, # data source 
           markers = ["o", "x"], 
           palette = "Set2") # 调色盘) #showing means and confidence intervals for unique values
```

```python
# dissolve - e.g. 根据城市名，把所有碎片拼起来

zhejiang_bycity = zhejiang.dissolve(by = 'CityNameC')
zhejiang_bycity
```

```python
# 查看整个shape file
zhejiang_bycity.plot(color='green')
plt.annotate(r"Zhejiang province", 
             xy=(1600000, 3000000),
             size = 10)
fig.tight_layout()
plt.show()
```

```python
data_list = [['杭州市', 18109, '8.5%'],
                                      ['宁波市', 14594.9, '8.2%'],
                                      ['温州市', 7585, '7.7%'],
                                      ['湖州市', 3644.9, '9.5%'],
                                      ['嘉兴市', 6355.28, '8.5%'],
                                      ['绍兴市', 6795, '8.7%'],
                                      ['金华市', 5355.44, '9.8%'],
                                      ['衢州市', 1875.61, '8.7%'],
                                      ['舟山市', 1703.62, '8.4%'],
                                      ['台州市', 5786.19, '8.3%'],
                                      ['丽水市', 1710.03, '8.3%']]

data_gdp_increase = pd.DataFrame(data_list, columns = ['CityNameC', 'gdp', 'increase'])
data_gdp_increase 
                                  
```

```python
# 如何合并两个的dataset？
print('length of dataset zhejiang:', len(zhejiang_bycity), '\nlength of dataset data_gdp_increase:', len(data_gdp_increase))

zhejiang_bycity = zhejiang_bycity.merge(data_gdp_increase, on='CityNameC')
zhejiang_bycity # 合成出来的表

# 把两个列表里的内容在了一个表里

```

```python
# plot zhejiang

fig, ax = plt.subplots(1, figsize=(10, 10))

zhejiang_bycity.plot(
    column = 'gdp', # 图例上按照颜色显示的数据,'increase'
    cmap = 'YlGnBu', # https://matplotlib.org/stable/tutorials/colors/colormaps.html # 颜色使用
    scheme ='NaturalBreaks', # 'Percentiles'/'JenksCaspall' Classification Schemes for Choropleth Maps 如何匹配数值和颜色
    linewidth = 1, 
    edgecolor = '0.25', # 0?1?
    ax=ax,
    legend = True)
ax.axis('off') # 这行删掉？


# add legend
leg = ax.get_legend()
leg.set_bbox_to_anchor((1.15,0.5))
leg.set_title('gdp\n')

plt.show()


```

```python
fig, ax = plt.subplots(1, figsize=(10, 10))

zhejiang_bycity.plot(
    column = 'gdp', 
    cmap = 'autumn', # https://matplotlib.org/stable/tutorials/colors/colormaps.html # 颜色使用
    scheme ='JenksCaspall', # 'Percentiles'/'JenksCaspall' Classification Schemes for Choropleth Maps 如何匹配数值和颜色
    linewidth = 1, 
    edgecolor = 'grey', 
    ax=ax,
    legend = True,
    legend_kwds = {'borderpad': 0.25, 
                   'fontsize':9}) 

    # 我们正在使用 geopandas 作图, 在 bing.com 检索 geopandas dataframe plot
    # -> 打开 geopandas 的官方文件，看看里面有没有提到 legend - legend_kwds
    # https://geopandas.org/en/stable/docs/reference/api/geopandas.GeoDataFrame.plot.html
    # 看到 legend_kwds的部分写道，Keyword arguments to pass to matplotlib.pyplot.legend()
    # 所以可以查找 matplotlib.pyplot.legend() 的官方文档 https://matplotlib.org/stable/api/legend_api.html
    # 关键词检索 size：-> borderpad -> The fractional whitespace inside the legend border, in font-size units. 
    # 也可以通过font-size调整：fontsizeint or {'xx-small', 'x-small', 'small', 'medium', 'large', 'x-large', 'xx-large'}
    
    
#ax.axis('off')


# 接下来要做什么 - 要把对应的区域填充进去

for index, row in zhejiang_bycity.iterrows():# 遍历每一行
    xy = row['geometry'].centroid.coords[:]
    print(row['CityNameC'], '的中心点为：', xy[0])
    ax.annotate(row['CityNameC'], # 具体annotation的内容：如xx市
                xy=xy[0], # annotation的位置
                horizontalalignment='center', # 水平对齐
                verticalalignment='center', # 垂直对齐
                fontproperties=zhfont,
                color = 'black',
                size = 13)
    
ax.set_title('浙江省2021年gdp分布', 
             fontproperties=zhfont, size = 16, color = 'red', weight = 'bold')

# add legend 可以理解成提取+移动 legend 的过程
leg = ax.get_legend() # get legend
leg.set_bbox_to_anchor((0.26,1)) # 设置legend坐标
leg.set_title('GDP')


plt.show()

```
