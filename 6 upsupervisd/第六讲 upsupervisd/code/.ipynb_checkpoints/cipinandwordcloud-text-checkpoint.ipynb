```python
data = pd.read_csv('Huxijin.csv')
data
```

```python
#r = requests.get('https://raw.githubusercontent.com/goto456/stopwords/master/%E5%93%88%E5%B7%A5%E5%A4%A7%E5%81%9C%E7%94%A8%E8%AF%8D%E8%A1%A8.txt')
#stopwords = r.text.split()
stopwords = [ line.rstrip().decode('gbk').encode('utf8') for line in open('chineseStopWords.txt') ]
stopwords[:10]
```

```python
def cut(texts):
    txt_list = []
    cixing=["x","zg","uj","ul","e","d","uz","y"]
    for i, txt in enumerate(texts):
        result = ''
        for w in pseg.cut(txt):
            if str(w.flag)  not in cixing:
                seg=w.word
                if seg not in stopwords:
                    result += (' '+seg)
        txt_list.append(result)
    return txt_list
```

```python
data['Content'] = cut(data['Content'].fillna(''))
```

```python
title_join = ' '.join(data['Content'])
title_counts = collections.Counter(title_join.split())
pd.DataFrame.from_dict(title_counts, orient='index').to_excel('Huxijin_freq.xlsx')
```
